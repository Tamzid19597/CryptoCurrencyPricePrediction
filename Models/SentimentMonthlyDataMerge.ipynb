```python
import pandas as pd 
Jan=pd.read_csv('january.csv',error_bad_lines=False,engine = 'python',header = None)
Feb=pd.read_csv('february.csv',error_bad_lines=False,engine = 'python',header = None)
Mar=pd.read_csv('march.csv',error_bad_lines=False,engine = 'python',header = None)
Apr=pd.read_csv('april.csv',error_bad_lines=False,engine = 'python',header = None)
May=pd.read_csv('may.csv',error_bad_lines=False,engine = 'python',header = None)
Jun=pd.read_csv('june.csv',error_bad_lines=False,engine = 'python',header = None)
Jul=pd.read_csv('july.csv',error_bad_lines=False,engine = 'python',header = None)
Aug=pd.read_csv('august.csv',error_bad_lines=False,engine = 'python',header = None)
Sep=pd.read_csv('september.csv',error_bad_lines=False,engine = 'python',header = None)
Oct=pd.read_csv('october.csv',error_bad_lines=False,engine = 'python',header = None)
Nov=pd.read_csv('november.csv',error_bad_lines=False,engine = 'python',header = None)
Final_2022=pd.concat([Jan,Feb,Mar,Apr,May,Jun,Jul,Aug,Sep,Oct,Nov])
Final_2022.to_csv('Sentiment2022.csv')
```
